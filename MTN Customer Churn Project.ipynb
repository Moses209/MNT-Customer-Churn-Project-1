```python
# To calculate the percentage of customers using the product per state
# First, we need to count unique customers per state (since some Customer IDs appear multiple times)
customers_per_state = df.drop_duplicates('Customer ID').groupby('State')['Customer ID'].count()
```

```python
# Calculate the percentage
percentage_per_state = (customers_per_state / customers_per_state.sum()) * 100
```

```python
# Sort for better visualization (optional)
percentage_per_state = percentage_per_state.sort_values(ascending=False)
```

```python
# Display the result
print("Percentage of customers per state:")
print(percentage_per_state.round(2).astype(str) + '%')
```

```python
# To see both count and percentage in a nice table:
result = pd.DataFrame({
    'Customer Count': customers_per_state,
    'Percentage': percentage_per_state.round(2)
})
print("\nDetailed breakdown:")
print(result)
```

```python
# Calculate the total sum of Customer Count
total_customers = result['Customer Count'].sum()

# Add a 'Total' row to the DataFrame
result.loc['Total'] = [total_customers, 100.00]  # Percentage total is 100%

# Display the result with the total
print("\nDetailed breakdown with total:")
print(result.to_markdown(floatfmt=".2f"))
```
